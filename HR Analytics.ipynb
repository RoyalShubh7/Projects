```python

# data cleaning:



df = df.drop(["Over18"],axis=1)
df = df.drop(["StandardHours"],axis=1)
df = df.drop(["EmployeeNumber"],axis=1)
df = df.drop(["EmployeeCount"],axis=1)
```

```python
# data processing :

# using the label encoder technique:

from sklearn.preprocessing import LabelEncoder

for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column]= LabelEncoder().fit_transform(df[column])
        
    
```

```python
df.dtypes
```

```python
# split the data:

x=df.drop("Attrition",axis=1)



y=df["Attrition"]
```

```python
x
```

```python
y
```

```python
# Split data in 80% and 20%:
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split

    
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state = 0)
#     lr=LogisticRegression()

#     lr.fit(x_train,y_train)
#     predrf = lr.predict(x_test)
#     acc = accuracy_score(y_test,predrf)
#     print("Accuracy", maxAcc, "Random State", i)


# if acc>maxAcc:
#     maxAcc==acc
#     maxRs=i
#     print("Accuracy", maxAcc, "Random state", i)
```

```python
# using Random State Classifier:


from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()


rfc.fit(x_train,y_train)
```

```python
# finding accuracy on the training set:
# Testing accuracy


rfc.score(x_train, y_train)
```

```python
# here accuracy of our model is 98.04% which is good one.
```

```python
# Showing the confusiong matrix and Acuuracy score on test data:


matrix = confusion_matrix(y_test, rfc.predict(x_test))

matrix
```

```python
print(classification_report(y_test, rfc.predict(x_test)))
```

```python
# using logistic regression:

from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
lr=LogisticRegression()

lr.fit(x_train,y_train)
lr.score(x_train, y_train)
matrix = confusion_matrix(y_test, rfc.predict(x_test))

matrix

print(classification_report(y_test, rfc.predict(x_test)))

```

```python
# Cross Validation score:
# foor Random Forest Classifier
# validation accuracy:

from sklearn.model_selection import cross_val_score
cvs = cross_val_score(rfc, x, y, cv=19)
print("cross avlidation score of Randoom Forest Classifier Model : ", cvs.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# creating parameter list:
parameters = {'max_features' : ['auto', 'sqrt','log2'],
              'max_depth' : [4,5,6,7,8] ,
              'criterion' : ['gini', 'entropy']
             }
```

```python
GCV = GridSearchCV(RandomForestClassifier(),parameters,cv=19,scoring = 'accuracy')

GCV.fit(x_train, y_train)

GCV.best_params_
```

```python
type(GCV)
```

```python
GCV.best_estimator_
```

```python
# predicting test with best parameters:

GCV_pred = GCV.best_estimator_.predict(x_test)

# Checking the final accuracy:

accuracy_score(y_test, GCV_pred)
```

```python
# Plotting the AUC ROC Curve:

from sklearn.metrics import plot_roc_curve
plot_roc_curve(GCV.best_estimator_, x_test, y_test)
plt.title("ROC AUC Plot")
plt.show()
```
